**Rainfall prediction using GMM (GaussianMixture)**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score, f1_score, mean_squared_error, confusion_matrix
from sklearn.preprocessing import StandardScaler
from scipy.stats import mode

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/rainfall prediction using ml dataset.csv')

# Replace 'RF' with the actual column name for the target variable
X = data.drop(columns=['RF'])
y = data['RF']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Convert continuous target variable into binary labels (rain or no rain)
threshold = y.median()  # Using median as threshold for classification
y_train_bin = (y_train > threshold).astype(int)
y_test_bin = (y_test > threshold).astype(int)

# Train the Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X_train)

# Predict the probabilities and classify based on the highest probability
y_train_pred = gmm.predict(X_train)
y_test_pred = gmm.predict(X_test)

# Map GMM cluster labels to binary labels
labels_train = np.zeros_like(y_train_pred)
labels_test = np.zeros_like(y_test_pred)

for i in range(2):
    mask_train = (y_train_pred == i)
    mask_test = (y_test_pred == i)

    # Ensure non-empty clusters
    if np.any(mask_train):
        labels_train[mask_train] = mode(y_train_bin[mask_train], keepdims=True).mode[0]
    if np.any(mask_test):
        labels_test[mask_test] = mode(y_test_bin[mask_test], keepdims=True).mode[0]

# Classification Metrics
accuracy = accuracy_score(y_test_bin, labels_test)
f1 = f1_score(y_test_bin, labels_test, average='weighted')
conf_matrix = confusion_matrix(y_test_bin, labels_test)

# Regression Fix: Predict rainfall as the mean rainfall per cluster
rainfall_means = {i: y_train[y_train_pred == i].mean() for i in range(2)}
y_test_pred_rainfall = np.array([rainfall_means[label] for label in y_test_pred])

# Regression Metrics
mse = mean_squared_error(y_test, y_test_pred_rainfall)
rmse = np.sqrt(mse)

# Print evaluation metrics
print(f'Classification Metrics:')
print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Confusion Matrix:\n{conf_matrix}')

print('\nRegression Metrics:')
print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score, f1_score, mean_squared_error, confusion_matrix
from sklearn.preprocessing import StandardScaler
from scipy.stats import mode

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/rainfall prediction using ml dataset.csv')

# Preprocess the data
X = data.drop(columns=['RF'])
y = data['RF']

# Convert continuous target variable into binary labels (0 = No Rain, 1 = Rain)
threshold = y.median()  # Using median as a threshold for classification
y_bin = (y > threshold).astype(int)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_bin, test_size=0.3, random_state=42)

# Train the Decision Tree model
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)

# Train the Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X_train)

# Predict the target for the testing data using Decision Tree
y_test_pred_tree = decision_tree.predict(X_test)

# Predict the probabilities and classify based on the highest probability using GMM
y_test_pred_prob_gmm = gmm.predict_proba(X_test)
y_test_pred_gmm = np.argmax(y_test_pred_prob_gmm, axis=1)

# Ensure mode returns a single value for each pair
y_test_pred_combined = np.array([mode([tree, gmm], keepdims=True).mode[0] for tree, gmm in zip(y_test_pred_tree, y_test_pred_gmm)])

# Calculate accuracy
accuracy = accuracy_score(y_test, y_test_pred_combined)

# Calculate F1 score
f1 = f1_score(y_test, y_test_pred_combined, average='weighted')

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_test_pred_combined)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_test_pred_combined)

# Print the evaluation metrics
print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
print('Confusion Matrix:')
print(conf_matrix)
```
